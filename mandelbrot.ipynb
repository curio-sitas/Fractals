```python
import PIL
import numpy as np
import pylab as plt
from tqdm import tqdm
```

```python
Nx= 1000
Ny = Nx
nmax = 70
x0 = -0.5

dx=2
dy = dx*Ny/Nx
y0 = -0
x = x0+np.linspace(-dx,dx,Nx,dtype=np.float64)
y = y0+np.linspace(-dy, dy,Ny, dtype=np.float64)
S = np.zeros((Nx,Ny))

for i in range(len(x)):
    for j in range(len(y)):
        xx = x[i]
        yy = y[j]
        it = 0
        zx = 0
        zy = 0
        while zx*zx + zy*zy < 4 and (it < nmax):
            xtemp = zx*zx - zy*zy + xx
            zy = 2*zx*zy + yy
            zx = xtemp
            it = it + 1
        S[i,j] = it

```

```python
plt.figure(figsize=(10,10))
plt.imshow(S, cmap='inferno')
plt.show()
```

```python
plt.imsave('mandelbrot.png', S, cmap='inferno', dpi=600)
```
